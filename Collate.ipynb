```python
gameData["Spread"] = 0
gameData["Spread Exists"] = False
```

```python
for i in range(len(spreadData)):
  row = (spreadData.iloc[i])
  date = str(row['date'])[:4] + '-' + str(row['date'])[4:6] + '-' + str(row['date'])[6:]
  homeStart = row['home']
  homeStart = homeStart.replace("L.A.","Los Angeles")
  spread = row["home_line"]
  gameData.loc[(gameData['Date'] == date) & (gameData['Home'].str.startswith(homeStart)),"Spread"] = spread
  gameData.loc[(gameData['Date'] == date) & (gameData['Home'].str.startswith(homeStart)),"Spread Exists"] = True
```

```python
dataset = gameData[gameData["Spread Exists"] == True]
```

```python
dataset = dataset.drop(columns=["Unnamed: 0", "Spread Exists"])
```

```python
dataset['Home Differential'] = dataset.apply(lambda x: x["Home Points"] - x["Visitor Points"], axis=1)
```

```python
dataset.columns
```

```python
teams = list(dataset["Visitor"].unique())
```

```python
dataset=dataset.dropna()
```

```python
dataset['Date'] = dataset.apply(lambda x: datetime.strptime(x['Date'], "%Y-%m-%d"), axis=1)
```

```python
dataset.head()
```

```python
def getRowsInPastXDays(df, numDays):
  idxs = []
  for i in range(1,len(df)):
    rowIDX = []
    j = i-1
    date1 = df.iloc[i]['Date']
    date2 = df.iloc[j]['Date']
    while (date1 - date2).days <= numDays and j >= 0:
      rowIDX.append(j)
      j-=1
      date2 = df.iloc[j]['Date']
    idxs.append(rowIDX)
  return idxs
```

```python
def getMiles(df, indexes, team):
  rows = df.iloc[indexes]
  miles = 0
  for row in rows.iterrows():
    if row[1]['Home'] == team:
      miles += row[1]['Home Miles']
    else:
      miles += row[1]['Visitor Miles']
  return miles
```

```python
teamDfs = {}
for team in teams:
  homeDf = dataset[dataset['Home'] == team]
  homeDf['Home Status'] = 1
  visitorDf = dataset[dataset['Visitor'] == team]
  visitorDf['Home Status'] = 0
  teamDf = pd.concat([homeDf, visitorDf])
  teamDf = teamDf.sort_values(by=['Date']).reset_index()
  teamDf['Days from previous game'] = float('inf')
  teamDf['Home Miles in Last 7 Days'] = 0
  teamDf['Games in Last 7 Days'] = 0
  teamDf['Games in Last 5 Days'] = 0
  teamDf['Games in Last 3 Days'] = 0
  teamDf['Covered'] = 0
  rowsInPrevious7Days = getRowsInPastXDays(teamDf,7)
  rowsInPrevious3Days = getRowsInPastXDays(teamDf,3)
  rowsInPrevious5Days = getRowsInPastXDays(teamDf,5)
  for i in range(1,len(teamDf)):
    teamDf.loc[i,'Days from previous game'] = (teamDf['Date'].iloc[i] - teamDf['Date'].iloc[i-1]).days
    teamDf.loc[i, 'Home Miles in Last 7 Days'] = getMiles(teamDf, rowsInPrevious7Days[i-1], team)
    teamDf.loc[i, 'Games in Last 7 Days'] = len(rowsInPrevious7Days[i-1])
    teamDf.loc[i, 'Games in Last 3 Days'] = len(rowsInPrevious3Days[i-1])
    teamDf.loc[i, 'Games in Last 5 Days'] = len(rowsInPrevious5Days[i-1])
    # If (team is home team and they beat they spread) or (team is away team and they beat the spread)
    if (teamDf.loc[i, 'Home Status'] == 1 and teamDf.loc[i, 'Home Differential'] > -teamDf.loc[i, 'Spread']) or (teamDf.loc[i, 'Home Status'] == 0 and teamDf.loc[i, 'Home Differential'] < -teamDf.loc[i, 'Spread']):
      teamDf.loc[i, 'Covered'] = 1
  teamDf.dropna(subset = ['Home','Visitor'])
  #teamDf = teamDf.drop(columns = ['Visitor','Visitor Points','Home','Home Points', 'OT'])
  teamDfs[team]=teamDf
```
